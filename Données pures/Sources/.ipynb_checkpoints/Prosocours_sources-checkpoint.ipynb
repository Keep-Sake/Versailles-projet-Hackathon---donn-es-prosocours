### Création d'une base de données

###### Nettoyage de la base de données brut pour ne garder que les données sourcées

```python
import json

input_file = "../Downloads/lien-interpersonnel.rehydrate.json"
output_file = "liens_interpersonnels_with_source.json"
skipped_file = "liens_interpersonnels_skipped.json"  # all skipped

def has_valid_source(doc):
    source = doc.get("_source", {}).get("source")
    return source is not None and str(source).strip() != ""

objects = []
skipped = []

buffer = ""
brace_count = 0

with open(input_file, "r", encoding="utf-8") as f:
    for line in f:
        stripped = line.strip()
        if not stripped:
            continue

        brace_count += stripped.count("{")
        brace_count -= stripped.count("}")

        buffer += line

        if brace_count == 0 and buffer.strip():
            try:
                obj = json.loads(buffer)

                if has_valid_source(obj):
                    objects.append(obj)
                else:
                    skipped.append({
                        "reason": "empty_or_missing_source",
                        "object": obj
                    })

            except json.JSONDecodeError as e:
                skipped.append({
                    "reason": "invalid_json",
                    "error": str(e),
                    "raw_block": buffer
                })

            buffer = ""

# write kept records (same structure as input)
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(objects, f, ensure_ascii=False, indent=2)

# write all skipped records to a separate file
with open(skipped_file, "w", encoding="utf-8") as f:
    json.dump(skipped, f, ensure_ascii=False, indent=2)

print(f"Done.")
print(f"Kept records: {len(objects)}")
print(f"Skipped records: {len(skipped)}")
```

```python
import json

input_file = "../Downloads/lien-interpersonnel.rehydrate.json"
output_file = "Final_liens_interpersonnels_with_source.json"
skipped_file = "Final_liens_interpersonnels_skipped.json"  # all skipped

def has_valid_source(doc):
    source = doc.get("_source", {}).get("source")
    return source is not None and str(source).strip() != ""

objects = []
skipped = []

buffer = ""
brace_count = 0

with open(input_file, "r", encoding="utf-8") as f:
    for line in f:
        stripped = line.strip()
        if not stripped:
            continue

        brace_count += stripped.count("{")
        brace_count -= stripped.count("}")

        buffer += line

        if brace_count == 0 and buffer.strip():
            try:
                obj = json.loads(buffer)

                if has_valid_source(obj):
                    objects.append(obj)
                else:
                    skipped.append({
                        "reason": "empty_or_missing_source",
                        "object": obj
                    })

            except json.JSONDecodeError as e:
                skipped.append({
                    "reason": "invalid_json",
                    "error": str(e),
                    "raw_block": buffer
                })

            buffer = ""

# --- Write kept records as **one JSON object per line** ---
with open(output_file, "w", encoding="utf-8") as f:
    for obj in objects:
        f.write(json.dumps(obj, ensure_ascii=False) + "\n")

# --- Write all skipped records as one JSON object per line ---
with open(skipped_file, "w", encoding="utf-8") as f:
    for obj in skipped:
        f.write(json.dumps(obj, ensure_ascii=False) + "\n")

print(f"Done.")
print(f"Kept records: {len(objects)}")
print(f"Skipped records: {len(skipped)}")
```

### Exploitation des données sourcées

###### Création d'un CSV de statistiques

```python
## Tableau de données sur institutions
import json
import pandas as pd


# --- CONFIG ---
PERSON_FILE = "../Downloads/personne.rehydrate.json"


# --- STREAM JSON OBJECTS ---
def stream_json_objects(filepath):
    buffer = ""
    brace_count = 0

    with open(filepath, "r", encoding="utf-8") as f:
        for line in f:
            buffer += line
            brace_count += line.count("{")
            brace_count -= line.count("}")

            if brace_count == 0 and buffer.strip():
                try:
                    yield json.loads(buffer)
                except json.JSONDecodeError:
                    pass
                buffer = ""


# --- EXTRACT DEEPEST NIVEAU (FOOLPROOF) ---
def extract_highest_niveau(person):
    source = person.get("_source", person)
    denorm = source.get("denormalization")

    if not isinstance(denorm, dict):
        return None, None

    charges = denorm.get("charges")
    if not isinstance(charges, list):
        return None, None

    highest_level = None
    highest_institution = None
    has_niveau_01 = False

    for charge in charges:
        if not isinstance(charge, dict):
            continue

        charge_obj = charge.get("charge")
        if not isinstance(charge_obj, dict):
            continue

        inst = charge_obj.get("institution")
        if not isinstance(inst, dict):
            continue

        if isinstance(inst.get("niveau_01"), dict):
            has_niveau_01 = True

        for i in range(1, 8):
            niv = inst.get(f"niveau_{i:02d}")
            if isinstance(niv, dict) and niv.get("nom"):
                highest_level = i
                highest_institution = niv["nom"]

    if not has_niveau_01:
        return None, None

    return highest_level, highest_institution


# --- COLLECT PERSONS ---
rows = []

print("Analyse des personnes...")

for person in stream_json_objects(PERSON_FILE):
    niveau, institution = extract_highest_niveau(person)
    if not niveau or not institution:
        continue

    source = person.get("_source", person)

    sexe = source.get("sexe") or "inconnu"
    noble = source.get("noble") or "inconnu"
    log = source.get("avec_logement") or "inconnu"
    # normalize values
    sexe = sexe.lower()
    noble = noble.lower()
    log.lower()
    rows.append({
        "institution": institution,
        "niveau": niveau,
        "sexe": sexe,
        "noble": noble,
        "logement" : log
    })

df = pd.DataFrame(rows)
print(f"Personnes retenues : {len(df)}")


# --- AGGREGATION ---
def percent(series, value):
    return round((series == value).sum() / len(series) * 100, 2)


aggregated = []

for institution, g in df.groupby("institution"):
    total = len(g)

    aggregated.append({
        
        "institution": institution,
        "niveau": int(g["niveau"].max()),
        "Total": total,

        "homme (%)": percent(g["sexe"], "homme"),
        "femme (%)": percent(g["sexe"], "femme"),
        "sexe_inconnu (%)": round(
            100 - percent(g["sexe"], "homme") - percent(g["sexe"], "femme"), 2
        ),

        "Logé (%)": percent(g["logement"], "oui"),
        "Non-logé (%)": percent(g["logement"], "non"),
        "Logement inconnu":round(
            100 - percent(g["logement"], "oui") - percent(g["logement"], "non"), 2),
        

        "noble (%)": percent(g["noble"], "oui"),
        "non_noble (%)": percent(g["noble"], "non"),
        "noble_inconnu (%)": round(
            100 - percent(g["noble"], "oui") - percent(g["noble"], "non"), 2
            
        ),

       
    })


# --- FINAL DATAFRAME ---
df_final = pd.DataFrame(aggregated)

# sort by total descending
df_final = df_final.sort_values("Total", ascending=False)

print("\nAperçu final :")
print(df_final.head())


# --- EXPORT ---
df_final.to_csv("institutions_stats_percentages.csv", index=False)

print("\nCSV généré : institutions_stats_percentages.csv")
```

###### I - Réseaux institutions religieux sourcés

```python
import json
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict, Counter
import plotly.graph_objects as go
import plotly.express as px

PERSON_FILE = '../Downloads/personne.rehydrate.json'
LINK_FILE = '../Data/FINAL_liens_interpersonnels_with_source.json'

def stream_json_objectss(filepath, debug=False):
    """
    Yields JSON objects from a file where each line is a JSON object.
    debug=True prints the first 5000 chars of each object.
    """
    with open(filepath, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            line = line.strip()
            if not line:
                continue
            try:
                obj = json.loads(line)
                if debug:
                    raw = json.dumps(obj, ensure_ascii=False)
                    print(f"Item {i+1}: first 5000 chars:\n{raw[:5000]}\n{'-'*50}\n")
                yield obj
            except json.JSONDecodeError as e:
                print(f"JSON decode error on line {i+1}: {e}")
                continue


def extract_lien_type(source):
    """
    Extracts lien_type label from both raw and denormalized JSON.
    Returns lowercase string or "".
    """
    # Direct field
    lt = source.get('lien_type')
    if isinstance(lt, str):
        return lt.strip().lower()  # <--- strip spaces/newlines

    # Denormalized field
    denorm = source.get('denormalization', {})
    if isinstance(denorm, dict):
        lt_dn = denorm.get('lien_type')
        if isinstance(lt_dn, dict):
            label = lt_dn.get('label')
            if isinstance(label, str):
                return label.strip().lower()  # <--- strip here too

    return ""
# ---------- ROBUST JSON STREAMER ----------

def stream_json_objects(filepath):
    buffer = ""
    brace_count = 0

    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            buffer += line
            brace_count += line.count("{")
            brace_count -= line.count("}")

            if brace_count == 0 and buffer.strip():
                try:
                    yield json.loads(buffer)
                except json.JSONDecodeError:
                    pass
                buffer = ""


# ---------- HIERARCHY EXTRACTION ----------
def get_hierarchy_info(data):
    source = data.get('_source', data)
    denorm = source.get('denormalization', {})
    if not isinstance(denorm, dict):
        return None, None

    charges = denorm.get('charges', [])
    first_charge = None

    if isinstance(charges, list) and charges:
        first_charge = charges[0]
    elif isinstance(charges, dict):
        first_charge = charges

    if not isinstance(first_charge, dict):
        return None, None

    inst = first_charge.get('charge', {}).get('institution', {})
    if not isinstance(inst, dict):
        return None, None

    excluded = {
        "Sans institution définie",
        "Sans institution connue",
        "Inconnu",
        "Autres",
        "Sans institution"
    }

    deepest_lv = None
    branch_name = None

    for i in range(1, 8):
        niv = inst.get(f'niveau_{i:02d}')
        if isinstance(niv, dict):
            nom = niv.get('nom')
            if nom and nom not in excluded:
                deepest_lv = i
                branch_name = nom

    return (deepest_lv, branch_name) if branch_name else (None, None)


# ---------- SAFE PERSON ID EXTRACTION ----------
def extract_person_id(p):
    if isinstance(p, dict):
        return p.get('personne')
    if isinstance(p, str):
        return p
    return None


# ---------- 1. NETWORK CONSTRUCTION ----------
G = nx.Graph()
node_data = {}
level_branches = defaultdict(Counter)

print("Analyse de la hiérarchie...")
for data in stream_json_objects(PERSON_FILE):
    p_id = data.get('_id')
    lv, branch = get_hierarchy_info(data)
    if p_id and lv:
        node_data[p_id] = {'level': lv, 'branch': branch}
        level_branches[lv][branch] += 1
        G.add_node(p_id)

print(f"Personnes retenues : {G.number_of_nodes()}")

print("Extraction des liens religieux...")
edge_count = 0
#len(stream_json_objects(LINK_FILE));



for data in stream_json_objects(LINK_FILE):
    source = data.get('_source', data)
    lien_type = extract_lien_type(source)
    #print(lien_type)
   # print(lien_type)
    if "rel" not in lien_type:

        continue

    p1 = extract_person_id(source.get('personne_1'))
    p2 = extract_person_id(source.get('personne_2'))

    if p1 in G and p2 in G:
        G.add_edge(p1, p2)
        edge_count += 1

print(f"Liens ajoutés : {edge_count}")
G.remove_nodes_from(list(nx.isolates(G)))
print(f"Nœuds après nettoyage : {G.number_of_nodes()}")


# ---------- 2. POSITIONING (ORBITS) ----------
radii = {i: i * 35 for i in range(1, 8)}
pos = {}

branches_all = sorted({d['branch'] for d in node_data.values()})
color_map = plt.cm.get_cmap('tab20', len(branches_all))

for lv in range(1, 8):
    nodes_at_lv = [n for n in G.nodes() if node_data[n]['level'] == lv]
    if not nodes_at_lv:
        continue

    nodes_at_lv.sort(key=lambda n: node_data[n]['branch'])
    n_count = len(nodes_at_lv)

    for i, node in enumerate(nodes_at_lv):
        angle = 2 * np.pi * i / n_count
        r = radii[lv] + np.random.uniform(-3, 3)
        pos[node] = np.array([r * np.cos(angle), r * np.sin(angle)])


# ---------- 3. INTERACTIVE PLOTLY RENDER ----------

print("Préparation du rendu interactif Plotly...")

# ---- EDGES ----
edge_x = []
edge_y = []

for n1, n2 in G.edges():
    x0, y0 = pos[n1]
    x1, y1 = pos[n2]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

edge_trace = go.Scatter(
    x=edge_x,
    y=edge_y,
    mode="lines",
    line=dict(width=0.6, color="rgba(180,180,180,0.6)"),
    hoverinfo="none",
    name="Liens"
)

# ---- NODES (BY BRANCH) ----
node_traces = []

branches_all = sorted({d["branch"] for d in node_data.values() if d["branch"]})
colors = px.colors.qualitative.Alphabet + px.colors.qualitative.Dark24

for i, branch in enumerate(branches_all):
    nodes_in_branch = [
        n for n in G.nodes()
        if node_data[n]["branch"] == branch
    ]
    if not nodes_in_branch:
        continue

    x_coords = [pos[n][0] for n in nodes_in_branch]
    y_coords = [pos[n][1] for n in nodes_in_branch]

    hover_texts = [
        f"<b>{n}</b><br>"
        f"Institution: {branch}<br>"
        f"Niveau: {node_data[n]['level']}"
        for n in nodes_in_branch
    ]

    node_trace = go.Scatter(
        x=x_coords,
        y=y_coords,
        mode="markers",
        name=branch.upper() if "maison du roi" in branch.lower() else branch,
        marker=dict(
            size=6,
            color=colors[i % len(colors)],
            line=dict(width=0)
        ),
        text=hover_texts,
        hoverinfo="text"
    )

    node_traces.append(node_trace)

# ---- FIGURE ----
fig = go.Figure(data=[edge_trace] + node_traces)

# ---- ORBIT CIRCLES ----
for i in range(1, 8):
    fig.add_shape(
        type="circle",
        xref="x", yref="y",
        x0=-radii[i], y0=-radii[i],
        x1=radii[i], y1=radii[i],
        line=dict(color="#333", dash="dash", width=1)
    )

# ---- LAYOUT ----
fig.update_layout(
    title=dict(
        text="Réseaux religieux – Institutions",
        font=dict(size=26, color="white"),
        x=0.5
    ),
    paper_bgcolor="#080808",
    plot_bgcolor="#080808",
    showlegend=True,
    legend=dict(
        font=dict(color="white", size=11),
        itemsizing="constant",
        orientation="v",
        x=1.02,
        y=1
    ),
    margin=dict(l=20, r=20, t=80, b=20),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    hoverlabel=dict(
        bgcolor="#111",
        font_size=13,
        font_color="white"
    )
)

print("Génération du HTML interactif...")
fig.write_html("reseau_institutions_religieux_interactif.html")
print("✔ HTML généré : reseau_institutions_religieux_interactif.html")
```

###### III - Analyses des réseaux institutionnels

```python
import json
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict, Counter
import plotly.graph_objects as go
import plotly.express as px

PERSON_FILE = '../Downloads/personne.rehydrate.json'
LINK_FILE = '../Data/FINAL_liens_interpersonnels_with_source.json'

def stream_json_objectss(filepath, debug=False):
    """
    Yields JSON objects from a file where each line is a JSON object.
    debug=True prints the first 5000 chars of each object.
    """
    with open(filepath, 'r', encoding='utf-8') as f:
        #print("QAAAAAAAAAAAAAAAA")
        for i, line in enumerate(f):
            line = line.strip()
            if not line:
                continue
            try:
                obj = json.loads(line)
                if debug:
                    raw = json.dumps(obj, ensure_ascii=False)
                    print(f"Item {i+1}: first 5000 chars:\n{raw[:5000]}\n{'-'*50}\n")
                yield obj
            except json.JSONDecodeError as e:
                print(f"JSON decode error on line {i+1}: {e}")
                continue


def extract_lien_type(source):
    """
    Extracts lien_type label from both raw and denormalized JSON.
    Returns lowercase string or "".
    """
    # Direct field
    lt = source.get('lien_type')
    if isinstance(lt, str):
        return lt.strip().lower()  # <--- strip spaces/newlines

    # Denormalized field
    denorm = source.get('denormalization', {})
    if isinstance(denorm, dict):
        lt_dn = denorm.get('lien_type')
        if isinstance(lt_dn, dict):
            label = lt_dn.get('label')
            if isinstance(label, str):
                return label.strip().lower()  # <--- strip here too

    return ""
# ---------- ROBUST JSON STREAMER ----------

def stream_json_objects(filepath):
    buffer = ""
    brace_count = 0

    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            buffer += line
            brace_count += line.count("{")
            brace_count -= line.count("}")

            if brace_count == 0 and buffer.strip():
                try:
                    yield json.loads(buffer)
                except json.JSONDecodeError:
                    pass
                buffer = ""


# ---------- HIERARCHY EXTRACTION ----------
def get_hierarchy_info(data):
    source = data.get('_source', data)
    denorm = source.get('denormalization', {})
    if not isinstance(denorm, dict):
        return None, None

    charges = denorm.get('charges', [])
    first_charge = None

    if isinstance(charges, list) and charges:
        first_charge = charges[0]
    elif isinstance(charges, dict):
        first_charge = charges

    if not isinstance(first_charge, dict):
        return None, None

    inst = first_charge.get('charge', {}).get('institution', {})
    if not isinstance(inst, dict):
        return None, None

    excluded = {
        "Sans institution définie",
        "Sans institution connue",
        "Inconnu",
        "Autres",
        "Sans institution"
    }

    deepest_lv = None
    branch_name = None

    for i in range(1, 8):
        niv = inst.get(f'niveau_{i:02d}')
        if isinstance(niv, dict):
            nom = niv.get('nom')
            if nom and nom not in excluded:
                deepest_lv = i
                branch_name = nom

    return (deepest_lv, branch_name) if branch_name else (None, None)


# ---------- SAFE PERSON ID EXTRACTION ----------
def extract_person_id(p):
    if isinstance(p, dict):
        return p.get('personne')
    if isinstance(p, str):
        return p
    return None


# ---------- 1. NETWORK CONSTRUCTION ----------
G = nx.Graph()
node_data = {}
level_branches = defaultdict(Counter)

print("Analyse de la hiérarchie...")
for data in stream_json_objects(PERSON_FILE):
    p_id = data.get('_id')
    lv, branch = get_hierarchy_info(data)
    if p_id and lv:
        node_data[p_id] = {'level': lv, 'branch': branch}
        level_branches[lv][branch] += 1
        G.add_node(p_id)

print(f"Personnes retenues : {G.number_of_nodes()}")

print("Extraction des liens religieux...")
edge_count = 0
#len(stream_json_objects(LINK_FILE));



for data in stream_json_objects(LINK_FILE):
    source = data.get('_source', data)
    lien_type = extract_lien_type(source)
    #print(lien_type)
   # print(lien_type)
    if " " not in lien_type:

        continue

    p1 = extract_person_id(source.get('personne_1'))
    p2 = extract_person_id(source.get('personne_2'))

    if p1 in G and p2 in G:
        G.add_edge(p1, p2)
        edge_count += 1

print(f"Liens ajoutés : {edge_count}")
G.remove_nodes_from(list(nx.isolates(G)))
print(f"Nœuds après nettoyage : {G.number_of_nodes()}")


# ---------- 2. POSITIONING (ORBITS) ----------
radii = {i: i * 35 for i in range(1, 8)}
pos = {}

branches_all = sorted({d['branch'] for d in node_data.values()})
color_map = plt.cm.get_cmap('tab20', len(branches_all))

for lv in range(1, 8):
    nodes_at_lv = [n for n in G.nodes() if node_data[n]['level'] == lv]
    if not nodes_at_lv:
        continue

    nodes_at_lv.sort(key=lambda n: node_data[n]['branch'])
    n_count = len(nodes_at_lv)

    for i, node in enumerate(nodes_at_lv):
        angle = 2 * np.pi * i / n_count
        r = radii[lv] + np.random.uniform(-3, 3)
        pos[node] = np.array([r * np.cos(angle), r * np.sin(angle)])


# ---------- 3. INTERACTIVE PLOTLY RENDER ----------

print("Préparation du rendu interactif Plotly...")

# ---- EDGES ----
edge_x = []
edge_y = []

for n1, n2 in G.edges():
    x0, y0 = pos[n1]
    x1, y1 = pos[n2]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

edge_trace = go.Scatter(
    x=edge_x,
    y=edge_y,
    mode="lines",
    line=dict(width=0.6, color="rgba(180,180,180,0.6)"),
    hoverinfo="none",
    name="Liens"
)

# ---- NODES (BY BRANCH) ----
node_traces = []

branches_all = sorted({d["branch"] for d in node_data.values() if d["branch"]})
colors = px.colors.qualitative.Alphabet + px.colors.qualitative.Dark24

for i, branch in enumerate(branches_all):
    nodes_in_branch = [
        n for n in G.nodes()
        if node_data[n]["branch"] == branch
    ]
    if not nodes_in_branch:
        continue

    x_coords = [pos[n][0] for n in nodes_in_branch]
    y_coords = [pos[n][1] for n in nodes_in_branch]

    hover_texts = [
        f"<b>{n}</b><br>"
        f"Institution: {branch}<br>"
        f"Niveau: {node_data[n]['level']}"
        for n in nodes_in_branch
    ]

    node_trace = go.Scatter(
        x=x_coords,
        y=y_coords,
        mode="markers",
        name=branch.upper() if "maison du roi" in branch.lower() else branch,
        marker=dict(
            size=6,
            color=colors[i % len(colors)],
            line=dict(width=0)
        ),
        text=hover_texts,
        hoverinfo="text"
    )

    node_traces.append(node_trace)

# ---- FIGURE ----
fig = go.Figure(data=[edge_trace] + node_traces)

# ---- ORBIT CIRCLES ----
for i in range(1, 8):
    fig.add_shape(
        type="circle",
        xref="x", yref="y",
        x0=-radii[i], y0=-radii[i],
        x1=radii[i], y1=radii[i],
        line=dict(color="#333", dash="dash", width=1)
    )

# ---- LAYOUT ----
fig.update_layout(
    title=dict(
        text="Réseaux – Institutions",
        font=dict(size=26, color="white"),
        x=0.5
    ),
    paper_bgcolor="#080808",
    plot_bgcolor="#080808",
    showlegend=True,
    legend=dict(
        font=dict(color="white", size=11),
        itemsizing="constant",
        orientation="v",
        x=1.02,
        y=1
    ),
    margin=dict(l=20, r=20, t=80, b=20),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    hoverlabel=dict(
        bgcolor="#111",
        font_size=13,
        font_color="white"
    )
)

print("Génération du HTML interactif...")
fig.write_html("reseau_institutions_interactif.html")
print("✔ HTML généré : reseau_institutions_interactif.html")
```

###### IV - Bulles des liens religieux

```python
import json
import networkx as nx
import numpy as np
import plotly.graph_objects as go
from community import community_louvain
from collections import defaultdict
import plotly.express as px

# ============================================================
# 1. CONFIGURATION
# ============================================================
PERSON_FILE = '../Downloads/personne.rehydrate.json'
LINK_FILE = '../Data/FINAL_liens_interpersonnels_with_source.json'

def stream_json_objects(filepath):
    buffer = ""
    brace_count = 0

    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            buffer += line
            brace_count += line.count("{")
            brace_count -= line.count("}")

            if brace_count == 0 and buffer.strip():
                try:
                    yield json.loads(buffer)
                except json.JSONDecodeError:
                    pass
                buffer = ""
# ============================================================
# 2. EXTRACTION ET AGGREGATION PAR FAMILLE
# ============================================================
G = nx.Graph()
person_to_family = {}
family_stats = defaultdict(lambda: {'count': 0, 'levels': []})

print("Étape 1 : Analyse des familles...")
for data in stream_json_objects(PERSON_FILE):
    p_id = data.get('_id')
    source = data.get('_source') or {}
    
    noms = source.get('noms') or []
    name_parts = [n.get('nom') if isinstance(n, dict) else n for n in noms]
    valid_names = [n.strip() for n in name_parts if n and len(str(n).strip()) > 1]
    
    fam_key = valid_names[0] if valid_names else (source.get('affichage') or "Inconnu")
    if "(" in fam_key: fam_key = fam_key.split("(")[0].strip()
    
    denorm = source.get('denormalization') or {}
    charges = denorm.get('charges') or []
    first_charge = charges[0] if isinstance(charges, list) and len(charges) > 0 else {}
    inst = (first_charge.get('charge') or {}).get('institution') or {} if isinstance(first_charge, dict) else {}
    
    lv = 1
    if isinstance(inst, dict):
        for i in range(1, 8):
            niv_obj = inst.get(f'niveau_{i:02d}') or {}
            if isinstance(niv_obj, dict) and niv_obj.get('nom'):
                lv = i

    person_to_family[p_id] = fam_key
    family_stats[fam_key]['count'] += 1
    family_stats[fam_key]['levels'].append(lv)

# ============================================================
# 3. CONSTRUCTION DES LIENS (MIXTE)
# ============================================================
print("Étape 2 : Analyse des alliances...")
types_autorises = ["Lien(s) familiaux", "Lien(s) religieux"]

for data in stream_json_objects(LINK_FILE):
    source = data.get('_source') or {}
    if source.get('lien_type') in types_autorises:
        p1 = (source.get('personne_1') or {}).get('personne')
        p2 = (source.get('personne_2') or {}).get('personne')
        f1, f2 = person_to_family.get(p1), person_to_family.get(p2)

        if f1 and f2 and f1 != f2:
            if G.has_edge(f1, f2):
                G[f1][f2]['weight'] += 1
            else:
                G.add_edge(f1, f2, weight=1)

G.remove_nodes_from(list(nx.isolates(G)))
for node in G.nodes():
    G.nodes[node]['size'] = family_stats[node]['count']
    G.nodes[node]['influence'] = np.mean(family_stats[node]['levels'])

# ============================================================
# 4. CALCUL DU LAYOUT ET COMMUNAUTÉS
# ============================================================
print("Étape 3 : Calcul du layout...")
top_nodes = sorted(dict(G.degree(weight='weight')).items(), key=lambda x: x[1], reverse=True)[:200]
sub_g = G.subgraph([n for n, _ in top_nodes]).copy()

partition = community_louvain.best_partition(sub_g)
pos = nx.spring_layout(sub_g, k=0.35, iterations=150, seed=42)

# ============================================================
# 5. CONSTRUCTION DU GRAPHE PLOTLY
# ============================================================

# 5.1 Les Arêtes (Lignes) - Unifiée pour la performance
edge_x, edge_y = [], []
for u, v in sub_g.edges():
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#444'),
    hoverinfo='none',
    mode='lines',
    showlegend=False # On ne veut pas de "liens" dans la légende
)

# 5.2 Les Nœuds (Une trace par famille pour la légende)
node_traces = []
colors_palette = px.colors.qualitative.Prism + px.colors.qualitative.Pastel + px.colors.qualitative.Vivid

# On trie les familles par clan pour grouper visuellement la légende
sorted_families = sorted(sub_g.nodes(), key=lambda n: (partition[n], n))

for family in sorted_families:
    x, y = pos[family]
    comm_id = partition[family]
    size = np.sqrt(sub_g.nodes[family]['size']) * 15
    influence = sub_g.nodes[family]['influence']
    
    node_trace = go.Scatter(
        x=[x], y=[y],
        mode='markers',
        name=family, # Affiche le nom de la famille dans la légende
        legendgroup=f"Clan {comm_id}", # Permet de les associer par clan
        marker=dict(
            size=[size],
            color=colors_palette[comm_id % len(colors_palette)],
            line=dict(width=1, color='white'),
            opacity=0.9
        ),
        text=f"<b>Famille : {family}</b><br>Membres : {sub_g.nodes[family]['size']}<br>Influence : {influence:.1f}",
        hoverinfo='text'
    )
    node_traces.append(node_trace)

# 5.3 Mise en page
fig = go.Figure(data=[edge_trace] + node_traces)

fig.update_layout(
    title=dict(
        text="Réseaux : liens familiaux et religieux",
        font=dict(size=22, color='white'),
        x=0.5
    ),
    paper_bgcolor='#050505',
    plot_bgcolor='#050505',
    showlegend=True,
    legend=dict(
        title=dict(text="Liste des Familles", font=dict(color="gray")),
        font=dict(color="white", size=9),
        itemsizing='constant',
        orientation="v",
        x=1.02, y=1,
        tracegroupgap=2 # Espace entre les groupes de clans
    ),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    margin=dict(l=10, r=200, t=80, b=10), # Marge à droite pour la légende
    hoverlabel=dict(bgcolor="#111", font_size=13, font_family="Arial")
)

# Sauvegarde et affichage
print("Génération du fichier HTML...")
fig.write_html("bulles_familles__religieux_sourcé.html")
fig.show()
```
