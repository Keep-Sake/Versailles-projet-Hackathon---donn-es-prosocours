### Création de documents uniquement avec des données datées

```python
import json

def correction_dates_vie(json_entree, json_sortie, json_skip):
    corrected_docs = []
    skipped_docs = []
    buffer = ""
    brace_count = 0

    with open(json_entree, "r", encoding="utf-8") as f:
        for line in f:
            stripped = line.strip()
            if not stripped:
                continue

            # compter les accolades pour détecter un objet complet
            brace_count += stripped.count("{")
            brace_count -= stripped.count("}")

            buffer += line

            if brace_count == 0 and buffer.strip():
                try:
                    doc = json.loads(buffer)
                    source = doc.get("_source", {})
                    naissance = source.get("naissance", {}).get("date", {}).get("date")
                    mort = source.get("mort", {}).get("date", {}).get("date")
                    if naissance is not None or mort is not None:
                        corrected_docs.append(doc)
                    else:
                        skipped_docs.append(doc)
                except json.JSONDecodeError as e:
                    print("Skipped invalid object (JSON error):", e)
                    skipped_docs.append(buffer)  # on garde le texte brut
                buffer = ""

    # écrire le fichier corrigé
    if corrected_docs:
        with open(json_sortie, "w", encoding="utf-8") as d:
            for docs in corrected_docs:
                json.dump(docs, d, ensure_ascii=False, indent=2)
        print(f"Json créé :) Kept {len(corrected_docs)} records.")
    else:
        print("json problématique:/ Aucun objet valide trouvé.")

    # écrire le fichier des objets skipped
    if skipped_docs:
        with open(json_skip, "w", encoding="utf-8") as d:
            json.dump(skipped_docs, d, ensure_ascii=False, indent=2)
        print(f"Skipped objects saved: {len(skipped_docs)} records.")
    else:
        print("Aucun objet skipped.")

# utilisation
correction_dates_vie(
    "personne.rehydrate.json",
    "personnes_update_jan2026.json",
    "personnes_skipped.json"
)
```

```python
import json

##LES LIENS
def correction_dates_liens(json_entree, json_sortie, json_skip):
    def has_any_date(obj):
        """Vérifie récursivement si une clé contient 'date' et non vide."""
        if isinstance(obj, dict):
            for k, v in obj.items():
                if "date" in k.lower() and v is not None and str(v).strip() != "":
                    return True
                if has_any_date(v):
                    return True
        elif isinstance(obj, list):
            for item in obj:
                if has_any_date(item):
                    return True
        return False
    corrected_docs = []
    skipped_docs = []
    buffer = ""

    with open(json_entree, "r", encoding="utf-8") as f:
        for line in f:
            buffer += line
            buffer_strip = buffer.strip()
            if not buffer_strip:
                continue
            try:
                doc = json.loads(buffer_strip)
                if has_any_date(doc.get("_source", {})):
                    corrected_docs.append(doc)
                else:
                    skipped_docs.append(doc)
                buffer = ""  

            except json.JSONDecodeError:
                continue

    with open(json_sortie, "w", encoding="utf-8") as f:
        for docs in corrected_docs:
            json.dump(docs, f, ensure_ascii=False, indent=2)
    with open(json_skip, "w", encoding="utf-8") as f:
        json.dump(skipped_docs, f, ensure_ascii=False, indent=2)

    print(f"{len(corrected_docs)} documents conservés, {len(skipped_docs)} skipped")
correction_dates_liens(
    "lien_interpersonnel.rehydrate.json",
    "liens_update_jan2026.json",
    "liens_skipped.json"
)
```

###### I - Un graphique de tous les individus de la base de données 'personnes', et les reliant selon leur nom de famille

```python
import json
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

PERSON_FILE = 'personnes_update_jan2026.json'
LINK_FILE = 'liens_update_jan2026.json'

import json
import json
import re

def stream_json_objectss(filepath):
    buffer = ""
    brace_count = 0

    with open(filepath, 'r', encoding='utf-8') as f:
        for liney in f:
            for line in liney:
                buffer += line
                brace_count += line.count("{")
                brace_count -= line.count("}")
    
                if brace_count == 0 and buffer.strip():
                    try:
                        yield json.loads(buffer)
                    except json.JSONDecodeError:
                        pass
                    buffer = ""


def get_hierarchy_info(person):
    source = person.get('_source') or {}
    denorm = source.get('denormalization') or {}
    charges = denorm.get('charges') or []

    if not charges or not isinstance(charges, list):
        return 1, "Inconnu"

    charge0 = charges[0] or {}
    charge = charge0.get('charge')

    if not isinstance(charge, dict):
        return 1, "Inconnu"

    inst = charge.get('institution')
    if not isinstance(inst, dict):
        return 1, "Inconnu"

    excluded = {"Inconnu", "Autres", "Sans institution"}
    deepest_lv = 1
    branch_name = None

    for i in range(1, 8):
        niv = inst.get(f'niveau_{i:02d}')
        if isinstance(niv, dict):
            nom = niv.get('nom')
            if isinstance(nom, str) and nom not in excluded:
                deepest_lv = i
                branch_name = nom

    return deepest_lv, branch_name or "Inconnu"

def extract_family_name(display_name):
    """Return first part of display name before '('"""
    #print(display_name)
    if not display_name: return "Inconnu"
    return display_name.split("(")[0].strip()

print("Processing persons...")
person_to_family = {}
family_info = {}
person_level_branch = {}

for person in stream_json_objectss(PERSON_FILE):
    source = person.get('_source', person)
    p_id = source.get('id') or source.get('_id')
    display = source.get('affichage')
    family = extract_family_name(display)
    lv, branch = get_hierarchy_info(person)
    
    person_to_family[p_id] = family
    person_level_branch[p_id] = (lv, branch)
    
    if family not in family_info:
        family_info[family] = {'count':0, 'levels':[]}
    family_info[family]['count'] += 1
    family_info[family]['levels'].append(lv)

print("Processing links...")
G = nx.Graph()

for fam, info in family_info.items():
    avg_lv = sum(info['levels']) / len(info['levels']) if info['levels'] else 1
    G.add_node(fam, size=info['count'], influence=avg_lv)

for link in stream_json_objectss(LINK_FILE):
    source = link.get('_source', link)
    if source.get('lien_type') != "Lien(s) familiaux":
        continue  
    
    p1 = source.get('personne_1', {}).get('personne')
    p2 = source.get('personne_2', {}).get('personne')
    f1 = person_to_family.get(p1)
    f2 = person_to_family.get(p2)
    if f1 and f2 and f1 != f2:
        if G.has_edge(f1, f2):
            G[f1][f2]['weight'] += 1
        else:
            G.add_edge(f1, f2, weight=1)

# Visualisation
print("Visualizing graph...")
plt.figure(figsize=(20, 14), facecolor='black')

pos = {}
radii = {i: i*50 for i in range(1, 8)}
families_by_lv = defaultdict(list)
for fam in G.nodes:
    lvl = round(G.nodes[fam]['influence'])
    families_by_lv[lvl].append(fam)

for lv, fams in families_by_lv.items():
    n = len(fams)
    for i, fam in enumerate(fams):
        angle = 2*np.pi*i/n
        r = radii[lv] + np.random.uniform(-5,5)
        pos[fam] = np.array([r*np.cos(angle), r*np.sin(angle)])

edge_weights = [d['weight'] for u,v,d in G.edges(data=True)]
nx.draw_networkx_edges(G, pos, width=[w*0.3 for w in edge_weights], alpha=0.3, edge_color="white")

node_sizes = [G.nodes[fam]['size']*100 for fam in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='cyan', alpha=0.8)

for fam, (x,y) in pos.items():
    plt.text(x, y, fam, fontsize=10, fontweight='bold', color='white', ha='center')

plt.title("Goupes de familles selon leur hiérarchie", color='white', fontsize=20)
plt.axis('off')
plt.tight_layout()
plt.show()
```

###### I - Un graphique de toutes les familles, selon le rang moyen des individus

```python
import json
import re
import numpy as np
import networkx as nx
import plotly.graph_objects as go
from collections import defaultdict

PERSON_FILE = 'personnes_update_jan2026.json'
LINK_FILE   = 'liens_update_jan2026.json'

def stream_json_objects(filepath):
    buffer = ""
    brace_count = 0

    with open(filepath, 'r', encoding='utf-8') as f:
        for liney in f:
            for line in liney:
                buffer += line
                brace_count += line.count("{")
                brace_count -= line.count("}")
    
                if brace_count == 0 and buffer.strip():
                    try:
                        yield json.loads(buffer)
                    except json.JSONDecodeError:
                        pass
                    buffer = ""

def extract_family_name(display_name):
    if not display_name:
        return "Inconnu"
    return display_name.split("(")[0].strip()

def get_hierarchy_info(person):
    source = person.get('_source') or {}
    denorm = source.get('denormalization') or {}
    charges = denorm.get('charges') or []

    if not isinstance(charges, list) or not charges:
        return 1

    charge = charges[0].get('charge')
    if not isinstance(charge, dict):
        return 1

    inst = charge.get('institution')
    if not isinstance(inst, dict):
        return 1

    excluded = {"Inconnu", "Autres", "Sans institution"}
    deepest = 1

    for i in range(1, 8):
        niv = inst.get(f'niveau_{i:02d}')
        if isinstance(niv, dict):
            nom = niv.get('nom')
            if isinstance(nom, str) and nom not in excluded:
                deepest = i

    return deepest

print("Processing persons...")
person_to_family = {}
family_levels = defaultdict(list)

for person in stream_json_objects(PERSON_FILE):
    src = person.get('_source', person)
    pid = person.get('_id') or src.get('id')
    family = extract_family_name(src.get('affichage'))
    level = get_hierarchy_info(person)

    if pid:
        person_to_family[pid] = family
        family_levels[family].append(level)

#Visualisation
print("Processing familial links...")
G = nx.Graph()

for fam, levels in family_levels.items():
    avg_level = sum(levels) / len(levels)
    G.add_node(
        fam,
        size=len(levels),
        influence=avg_level
    )

for link in stream_json_objects(LINK_FILE):
    src = link.get('_source', link)
    if src.get('lien_type') != "Lien(s) familiaux":
        continue

    p1 = src.get('personne_1', {}).get('personne')
    p2 = src.get('personne_2', {}).get('personne')

    f1 = person_to_family.get(p1)
    f2 = person_to_family.get(p2)

    if f1 and f2 and f1 != f2:
        if G.has_edge(f1, f2):
            G[f1][f2]['weight'] += 1
        else:
            G.add_edge(f1, f2, weight=1)

G.remove_nodes_from(list(nx.isolates(G)))

print("Computing positions...")
pos = {}
radii = {i: i * 50 for i in range(1, 8)}
families_by_lv = defaultdict(list)

for fam in G.nodes:
    lv = int(round(G.nodes[fam]['influence']))
    lv = max(1, min(7, lv))
    families_by_lv[lv].append(fam)

for lv, fams in families_by_lv.items():
    n = len(fams)
    for i, fam in enumerate(sorted(fams)):
        angle = 2 * np.pi * i / max(n, 1)
        r = radii[lv] + np.random.uniform(-6, 6)
        pos[fam] = (r * np.cos(angle), r * np.sin(angle))

print("Building HTML graph...")

# Edges
edge_x, edge_y = [], []
for u, v in G.edges():
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    edge_x += [x0, x1, None]
    edge_y += [y0, y1, None]

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    mode='lines',
    line=dict(width=0.6, color='rgba(180,180,180,0.5)'),
    hoverinfo='none'
)

node_x, node_y, node_sizes, node_colors, hovers = [], [], [], [], []
max_size = max(nx.get_node_attributes(G, 'size').values())

for fam in G.nodes:
    x, y = pos[fam]
    node_x.append(x)
    node_y.append(y)

    size = G.nodes[fam]['size']
    node_sizes.append(6 + 20 * size / max_size)
    node_colors.append(G.nodes[fam]['influence'])

    hovers.append(
        f"<b>{fam}</b><br>"
        f"Membres : {size}<br>"
        f"Niveau moyen : {G.nodes[fam]['influence']:.2f}"
    )

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers+text',
    text=list(G.nodes),
    textposition="top center",
    hovertext=hovers,
    hoverinfo='text',
    marker=dict(
        size=node_sizes,
        color=node_colors,
        colorscale='Viridis',
        showscale=True,
        line=dict(width=0)
    )
)

fig = go.Figure(data=[edge_trace, node_trace])

for i in range(1, 8):
    fig.add_shape(
        type="circle",
        xref="x", yref="y",
        x0=-radii[i], y0=-radii[i],
        x1=radii[i], y1=radii[i],
        line=dict(color="#444", dash="dash", width=1)
    )

fig.update_layout(
    title="Réseau familial – hiérarchie institutionnelle",
    paper_bgcolor="#080808",
    plot_bgcolor="#080808",
    showlegend=False,
    margin=dict(l=20, r=20, t=80, b=20),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    hoverlabel=dict(bgcolor="#111", font_color="white")
)

fig.write_html("reseau_familial_orbital.html")
print("✔ Graphe généré : reseau_familial_orbital.html")
```

```python
import json
import re
import numpy as np
import networkx as nx
import plotly.graph_objects as go
from collections import defaultdict
import plotly.express as px
from collections import Counter
import pandas as pd

PERSON_FILE = 'personnes_update_jan2026.json'
LINK_FILE   = 'liens_update_jan2026.json'

def stream_json_objects(filepath):
    buffer = ""
    brace_count = 0

    with open(filepath, 'r', encoding='utf-8') as f:
        for liney in f:
            for line in liney:
                buffer += line
                brace_count += line.count("{")
                brace_count -= line.count("}")
    
                if brace_count == 0 and buffer.strip():
                    try:
                        yield json.loads(buffer)
                    except json.JSONDecodeError:
                        pass
                    buffer = ""

def get_hierarchy_info(person):
    source = person.get('_source') or {}
    denorm = source.get('denormalization') or {}
    charges = denorm.get('charges') or []
    if not isinstance(charges, list) or not charges:
        return 1
    charge = charges[0].get('charge')
    if not isinstance(charge, dict):
        return 1
    inst = charge.get('institution')
    if not isinstance(inst, dict):
        return 1
    excluded = {"Inconnu", "Autres", "Sans institution"}
    deepest = 1
    for i in range(1, 8):
        niv = inst.get(f'niveau_{i:02d}')
        if isinstance(niv, dict):
            nom = niv.get('nom')
            if isinstance(nom, str) and nom not in excluded:
                deepest = i
    return deepest

def get_display_name(person):
    source = person.get('_source') or {}
    return source.get('affichage') or "Anonyme"

def extract_family_name(display_name):
    if not display_name:
        return "Inconnu"
    return display_name.split("(")[0].strip()

#Visualisation
print("Processing individuals...")
G = nx.Graph()
node_metadata = {}

for person in stream_json_objects(PERSON_FILE):
    source = person.get('_source', person)
    pid = person.get('_id') or source.get('id')
    if not pid:
        continue
    level = get_hierarchy_info(person)
    name = get_display_name(person)
    family = extract_family_name(name)
    G.add_node(pid)
    node_metadata[pid] = {'level': level, 'name': name, 'family': family}

print("Adding familial links...")
for link in stream_json_objects(LINK_FILE):
    source = link.get('_source', link)
    if source.get('lien_type') != "Lien(s) familiaux":
        continue
    p1 = source.get('personne_1', {}).get('personne')
    p2 = source.get('personne_2', {}).get('personne')
    if p1 in G and p2 in G:
        if G.has_edge(p1, p2):
            G[p1][p2]['weight'] += 1
        else:
            G.add_edge(p1, p2, weight=1)

print("Computing positions by level...")
radii = {i:i*50 for i in range(1,8)}
pos = {}
for lv in range(1,8):
    nodes_at_lv = [n for n in G.nodes() if node_metadata[n]['level']==lv]
    n_count = len(nodes_at_lv)
    for i, node in enumerate(nodes_at_lv):
        angle = 2*np.pi*i/max(n_count,1)
        r = radii[lv] + np.random.uniform(-5,5)
        pos[node] = (r*np.cos(angle), r*np.sin(angle))

print("Building interactive plot by family...")

edge_x, edge_y = [], []
for u,v in G.edges():
    x0,y0 = pos[u]
    x1,y1 = pos[v]
    edge_x += [x0,x1,None]
    edge_y += [y0,y1,None]

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.6,color='rgba(180,180,180,0.5)'),
    hoverinfo='none',
    mode='lines',
    name='Liens familiaux'
)

families = sorted(list(set(node_metadata[n]['family'] for n in G.nodes())))
colors = px.colors.qualitative.Alphabet + px.colors.qualitative.Dark24
node_traces = []

for i, fam in enumerate(families):
    nodes_in_fam = [n for n in G.nodes() if node_metadata[n]['family']==fam]
    x = [pos[n][0] for n in nodes_in_fam]
    y = [pos[n][1] for n in nodes_in_fam]
    hover_text = [node_metadata[n]['name'] for n in nodes_in_fam]
    node_trace = go.Scatter(
        x=x, y=y,
        mode='markers',
        marker=dict(
            size=6,
            color=colors[i % len(colors)],
            line=dict(width=0)
        ),
        text=hover_text,
        hoverinfo='text',
        name=fam,
        legendgroup=fam,
        showlegend=True
    )
    node_traces.append(node_trace)

fig = go.Figure(data=[edge_trace] + node_traces)

for i in range(1,8):
    fig.add_shape(
        type="circle",
        xref="x", yref="y",
        x0=-radii[i], y0=-radii[i],
        x1=radii[i], y1=radii[i],
        line=dict(color="#444", dash="dash", width=1)
    )

fig.update_layout(
    title="Réseau familial – tous les individus",
    paper_bgcolor="#080808",
    plot_bgcolor="#080808",
    showlegend=True,
    legend=dict(title="Familles", font=dict(color="white"), itemclick="toggle"),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    hoverlabel=dict(bgcolor="#111", font_color="white")
)


fig.write_html("reseau_familial_individus_par_famille.html")
print("✔ Graphe généré : reseau_familial_individus_par_famille.html, celui qui marche bien")


#Compter les familles et les répertorier
family_counts = Counter(node_metadata[n]['family'] for n in G.nodes())
total_families = len(family_counts)
total_persons = sum(family_counts.values())

df = pd.DataFrame({
    'Family': list(family_counts.keys()),
    'Members': list(family_counts.values())
})

df = df.sort_values('Members', ascending=False).reset_index(drop=True)

totals = pd.DataFrame({
    'Family': [f"Total families: {total_families}"],
    'Members': [total_persons]
})
df = pd.concat([totals, df], ignore_index=True)

df
```

```python
import json
import re
import numpy as np
import networkx as nx
import plotly.graph_objects as go
import plotly.express as px

print("HELLO")

PERSON_FILE = 'personnes_update_jan2026.json'
LINK_FILE = 'liens_update_jan2026.json'

def stream_json_objects(filepath):
    buffer = ""
    brace_count = 0
    with open(filepath, 'r', encoding='utf-8') as f:
        for liney in f:
            for char in liney:
                buffer += char
                brace_count += char.count("{")
                brace_count -= char.count("}")
                if brace_count == 0 and buffer.strip():
                    try:
                        yield json.loads(buffer)
                    except json.JSONDecodeError:
                        pass
                    buffer = ""

def extract_year(date_str):
    if not date_str:
        return None
    match = re.search(r'(\d{4})', str(date_str))
    return int(match.group(1)) if match else None

def get_birth_death(person):
    source = person.get('_source', {}) or {}
    birth, death = None, None

    naissance = source.get('naissance')
    if isinstance(naissance, dict):
        date_obj = naissance.get('date')
        if isinstance(date_obj, dict):
            birth = extract_year(date_obj.get('date'))
        elif isinstance(date_obj, str):
            birth = extract_year(date_obj)

    mort = source.get('mort')
    if isinstance(mort, dict):
        date_obj = mort.get('date')
        if isinstance(date_obj, dict):
            death = extract_year(date_obj.get('date'))
        elif isinstance(date_obj, str):
            death = extract_year(date_obj)

    if birth is None and death is not None:
        birth = death - 70
    if death is None and birth is not None:
        death = birth + 70

    return birth, death

def get_hierarchy_info(person):
    source = person.get('_source', {})
    denorm = source.get('denormalization', {}) or {}
    charges = denorm.get('charges') or []
    if not charges:
        return 1
    try:
        charge = charges[0].get('charge', {})
        inst = charge.get('institution', {})
        excluded = {"Inconnu", "Autres", "Sans institution"}
        deepest = 1
        for i in range(1, 8):
            niv = inst.get(f'niveau_{i:02d}') or {}
            nom = niv.get('nom')
            if isinstance(nom, str) and nom not in excluded:
                deepest = i
        return deepest
    except:
        return 1

def get_display_name(person):
    return (person.get('_source', {}) or {}).get('affichage', "Anonyme")

def extract_family_name(name):
    if not name:
        return "Inconnu"
    return name.split("(")[0].strip()

#Visualisation
print("Processing individuals...")
G = nx.Graph()
node_metadata = {}
persons_dict = {}

for person in stream_json_objects(PERSON_FILE):
    pid = person.get('_id') or person.get('_source', {}).get('id')
    if not pid:
        continue
    level = get_hierarchy_info(person)
    name = get_display_name(person)
    family = extract_family_name(name)
    G.add_node(pid)
    node_metadata[pid] = {'level': level, 'name': name, 'family': family}
    persons_dict[pid] = person

birth_years, death_years = [], []
for p in persons_dict.values():
    b, d = get_birth_death(p)
    if b: birth_years.append(b)
    if d: death_years.append(d)

MIN_YEAR = 1550
start_year = max(min(birth_years), MIN_YEAR)
end_year = max(death_years)
years = list(range(start_year, end_year + 1))

print(f"Timeline range: {start_year} → {end_year}")

print("Adding family links...")
for link in stream_json_objects(LINK_FILE):
    source = link.get('_source', link)
    if source.get('lien_type') != "Lien(s) familiaux":
        continue
    p1 = source.get('personne_1', {}).get('personne')
    p2 = source.get('personne_2', {}).get('personne')
    if p1 in G and p2 in G:
        if G.has_edge(p1, p2):
            G[p1][p2]['weight'] += 1
        else:
            G.add_edge(p1, p2, weight=1)

radii = {i: i * 50 for i in range(1, 8)}
pos = {}

for lv in range(1, 8):
    nodes_at_lv = [n for n in G.nodes() if node_metadata[n]['level'] == lv]
    for i, node in enumerate(nodes_at_lv):
        angle = 2 * np.pi * i / max(len(nodes_at_lv), 1)
        r = radii[lv] + np.random.uniform(-5, 5)
        pos[node] = (r * np.cos(angle), r * np.sin(angle))

print("Precomputing life ranges...")
life_range = {}
for pid, person in persons_dict.items():
    b, d = get_birth_death(person)
    if b is not None and d is not None:
        life_range[pid] = (b, d)

edge_life = []
for u, v in G.edges():
    if u in life_range and v in life_range:
        b = max(life_range[u][0], life_range[v][0])
        d = min(life_range[u][1], life_range[v][1])
        if b <= d:
            edge_life.append((u, v, b, d))

print("Building optimized timeline...")
colors = px.colors.qualitative.Alphabet + px.colors.qualitative.Dark24
families = sorted(set(node_metadata[n]['family'] for n in G.nodes()))

alive_nodes = set()
alive_edges = set()
frames = []

for year in years:
    print(year)

    for pid, (b, d) in life_range.items():
        if year == b:
            alive_nodes.add(pid)
        if year == d + 1:
            alive_nodes.discard(pid)

    for u, v, b, d in edge_life:
        if year == b:
            alive_edges.add((u, v))
        if year == d + 1:
            alive_edges.discard((u, v))

    edge_x, edge_y = [], []
    for u, v in alive_edges:
        x0, y0 = pos[u]
        x1, y1 = pos[v]
        edge_x += [x0, x1, None]
        edge_y += [y0, y1, None]

    edge_trace = go.Scatter(
        x=edge_x,
        y=edge_y,
        mode='lines',
        line=dict(width=0.6, color='rgba(180,180,180,0.5)'),
        hoverinfo='none'
    )

    node_traces = []
    for i, fam in enumerate(families):
        nodes = [n for n in alive_nodes if node_metadata[n]['family'] == fam]
        node_traces.append(go.Scatter(
            x=[pos[n][0] for n in nodes],
            y=[pos[n][1] for n in nodes],
            mode='markers',
            marker=dict(size=6, color=colors[i % len(colors)]),
            text=[node_metadata[n]['name'] for n in nodes],
            hoverinfo='text',
            name=fam,
            legendgroup=fam
        ))

    frames.append(go.Frame(name=str(year), data=[edge_trace] + node_traces))

sliders = [{
    "active": 0,
    "currentvalue": {"prefix": "Year: ", "font": {"size": 16, "color": "white"}},
    "pad": {"t": 50},
    "steps": [{
        "method": "animate",
        "args": [[str(year)], {"frame": {"duration": 0, "redraw": True}, "mode": "immediate"}],
        "label": str(year)
    } for year in years]
}]

fig = go.Figure(
    data=frames[0].data,
    layout=go.Layout(
        title="Évolution sociale et familiale",
        paper_bgcolor="#080808",
        plot_bgcolor="#080808",
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        hoverlabel=dict(bgcolor="#111", font_color="white"),
        sliders=sliders,
        updatemenus=[dict(
            type="buttons",
            showactive=False,
            buttons=[dict(
                label="Play",
                method="animate",
                args=[None, {"frame": {"duration": 10, "redraw": True}, "fromcurrent": True}]
            )]
        )]
    ),
    frames=frames
)

for i in range(1, 8):
    fig.add_shape(
        type="circle",
        xref="x", yref="y",
        x0=-radii[i], y0=-radii[i],
        x1=radii[i], y1=radii[i],
        line=dict(color="#444", dash="dash", width=1)
    )

fig.write_html("reseau_familial_interactive_timeline22.html")
print("✔ Graphe généré : reseau_familial_interactive_timeline22.html")
```
